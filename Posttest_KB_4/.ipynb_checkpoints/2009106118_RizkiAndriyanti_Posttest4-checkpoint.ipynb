## Cleaning Data

membersihkan data yang null, duplikat atau rusak

```python
dataset_1 = df.copy()
dataset_2 = df2.copy()
dataset_3 = df3.copy()
```

```python
input_mean = SimpleImputer(strategy='mean')
input_median = SimpleImputer(strategy='median')
input_modus = SimpleImputer(strategy='most_frequent')
```

## Modus

```python
dataset_1['Supervisor District'] = input_modus.fit_transform(dataset_1[['Supervisor District']])

dataset_1.isna().sum()
```

## Mean

```python
dataset_2[['Submission Year','Public Internet Computers','Wireless Internet Sessions']] = input_mean.fit_transform(dataset_2[['Submission Year','Public Internet Computers','Wireless Internet Sessions']])

dataset_2.isna().sum()
```

```python
dataset_onehot = df2.copy()
dataset_ordinal = df2.copy()
```

```python
dataset = pd.get_dummies(dataset_onehot[['Administrative Structure']])

dataset = pd.DataFrame(dataset)

dataset_onehot.drop('Administrative Structure', axis=1, inplace=True)

dataset_onehot = dataset_onehot.join(dataset)
dataset_onehot.head(10)
```
