## GARFIK BOXFLOT

```python
plt.boxplot(df[['Patron Type Code','Total Checkouts']], labels=['patron','total'])
plt.show()
```

## GRAFIK HEATMAP

```python
sns.heatmap(data=df.corr(), cmap = 'Blues')
plt.show()
```

## GRAFIK BATANG

jadi pada grafik batang pada atribut supervisor district memiliki nilai paling tinggi yaitu lebih dari 200.

```python
df['Supervisor District'].value_counts().plot.bar()
plt.show()
```

## GRAFIK HSTROGRAM

Pada grafik histogram, pada atribut supervisor district memiliki nilai paling tinggi yaitu lebih dari 200.

```python
plt.figure(figsize=(10,5))
plt.hist(df['Supervisor District'],label='supervisor',color='skyblue')
plt.legend()
plt.xticks(rotation =90)
plt.show()
```

## GRAFIK PIE

pada grafik pie, menjelaskan hasil sikulasi pada setiap bulan pada data set perpustakaan.

```python
plt.figure(figsize=(10,5))
plt.pie(df['Circulation Active Month'].value_counts(),labels=['Januari','Februari','Maret','April','Mei','Juni','Juli','Agustus','September','Oktober','November','Desember','None'], startangle=90, autopct='%.1f%%')
plt.legend(loc='lower right')
plt.show()
```

## GRAFIK PLOT DAN GARIS

```python
plt.figure(figsize=(10,5))
plt.title('Grafik Plot dan Garis')
plt.plot(df['Circulation Active Year'], df['Circulation Active Month'], linestyle ='--',color = 'skyblue')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

## SCATTER PLOT

```python
plt.figure(figsize=(10,5))
plt.title('Scatter Plot')
plt.scatter(df['Circulation Active Year'], df['Circulation Active Month'], linestyle ='--',color = 'skyblue')
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
